## looking at the heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test = [
    [1,0.5,0.5],
    [0.5,1,0.5],
    [0.5,0.5,1]
]
```

```python
plt.Figure(figsize=(160,80))
sns.heatmap(test, annot=True, cmap="viridis")
plt.show()
```

```python
plt.Figure(figsize=(160,80))
sns.heatmap(df.corr(), annot=True, cmap="viridis")
plt.show()
```
